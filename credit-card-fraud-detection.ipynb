## Handling imbalance data with three techniques
#### 1. Undersampling
#### 2. Oversampling
#### 3. SMOTE(Synthetic Minority oversampling technique)

## 1. Undersampling technique

```python
majority_class = df[df['Class']==0]
minority_class = df[df['Class']==1]
```

```python
majority_class.shape
```

```python
minority_class.shape
```

```python
majority_downsample = majority_class.sample(n=len(minority_class),replace=False)
```

```python
majority_downsample.shape
```

```python
undersample_balanced_df = pd.concat([majority_downsample,minority_class])
```

```python
undersample_balanced_df.shape
```

```python
undersample_balanced_df['Class'].value_counts()
```

```python
undersample_balanced_df.head()
```

```python
x = undersample_balanced_df.iloc[:,:-1]
y = undersample_balanced_df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=100)
```

```python
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

### Logistic regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
undersample_lr_prediction = lr.predict(X_test)
```

```python
from sklearn import metrics
```

```python
print(metrics.classification_report(y_test, undersample_lr_prediction))
```

### Decision tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
undersample_dt = DecisionTreeClassifier()
```

```python
undersample_dt.fit(X_train,y_train)
```

```python
undersample_dt_prediction = undersample_dt.predict(X_test)
```

```python
print(metrics.classification_report(y_test, undersample_dt_prediction))
```

### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
undersample_rf = RandomForestClassifier()
```

```python
undersample_rf.fit(X_train,y_train)
```

```python
undersample_rf_prediction = undersample_rf.predict(X_test)
```

```python
print(metrics.classification_report(y_test, undersample_rf_prediction))
```

### Visualizing Model accuracy scores

```python
Accuracy_scores=pd.DataFrame({'Models' : ['LR','DT','RF'],
                'Accuracy Scores':[metrics.accuracy_score(y_test, undersample_lr_prediction)*100,
                                   metrics.accuracy_score(y_test, undersample_dt_prediction)*100,
                                   metrics.accuracy_score(y_test, undersample_rf_prediction)*100]})
```

```python
Accuracy_scores
```

```python
sns.barplot(x=Accuracy_scores['Models'], y=Accuracy_scores['Accuracy Scores'], palette='viridis')
plt.xlabel('Models')
plt.ylabel('Accuracy Score')
plt.title('Accuracy Scores of Different Models after Undersampling technique')
plt.show()
```

### Random Forest Classifier has highest Accuracy after Under Sampling followed by Logistic regression and Decision tree

## 3. SMOTE(Synthetic minority Over-Sampling Technique)

```python
X = df.iloc[:,:-1]
Y = df.iloc[:,-1]
```

```python
print(X.shape)
print(Y.shape)
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE()
```

```python
X_resample, y_resample = smote.fit_resample(X,Y)
```

```python
print(X_resample.shape)
print(y_resample.shape)
```

```python
y_resample.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_resample,y_resample,test_size=0.30,random_state=100)
```

```python
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
log_reg = LogisticRegression()
```

```python
log_reg.fit(X_train,y_train)
```

```python
smote_lr_prediction = log_reg.predict(X_test)
```

```python
print(metrics.classification_report(y_test,smote_lr_prediction))
```

## Decision Tree

```python
d_t = DecisionTreeClassifier()
```

```python
d_t.fit(X_train,y_train)
```

```python
smote_dt_prediction = d_t.predict(X_test)
```

```python
print(metrics.classification_report(y_test,smote_dt_prediction))
```

## Random Forest

```python
R_F = RandomForestClassifier()
```

```python
R_F.fit(X_train,y_train)
```

```python
smote_rf_prediction = R_F.predict(X_test)
```

```python
print(metrics.classification_report(y_test,smote_rf_prediction))
```

### Visualizing model Accuracy Scores after SMOTE technique

```python
Accuracy_scores=pd.DataFrame({'Models' : ['LR','DT','RF'],
                'Accuracy Scores':[metrics.accuracy_score(y_test, smote_lr_prediction)*100,
                                   metrics.accuracy_score(y_test, smote_dt_prediction)*100,
                                   metrics.accuracy_score(y_test, smote_rf_prediction)*100]})
```

```python
Accuracy_scores
```

```python
sns.barplot(x=Accuracy_scores['Models'], y=Accuracy_scores['Accuracy Scores'], palette='dark')
plt.xlabel('Models')
plt.ylabel('Accuracy Score')
plt.title('Accuracy Scores of Different Models after SMOTE technique')
plt.show()
```

### Random forest and Decision tree both are having same accuracy after smote technique

### After applying all three techniques, Random Forest Classifier is the best model to perform on this dataset.
